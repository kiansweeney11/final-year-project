# Check for outliers

```python
plt.figure(figsize=[15,7])
sns.boxplot(x = df4['Price'], color=".25")
```

Here removing values over 2.5£ million seems a safe idea.

```python
df4 = df4[df4['Price'] < 2500000]
```

```python
df4.shape
```

We remove these outliers as like our Irish data these could potentially be whole apartment blocks or multiple flats and do not fall into the category of individual lettings in which our end goal is to predict the price of.

```python
dt = np.array(df4['Date of Transfer'])
month = []
for i in range(len(dt)):
    strg = dt[i].split('-')
    #print(strg)
    month.append(strg[1])
```

```python
month = np.array(month)
df4['Month'] = month.tolist()
```

```python
dt = np.array(df4['Date of Transfer'])
yr = []
for i in range(len(dt)):
    strg = dt[i].split('-')
    #print(strg)
    yr.append(strg[0])
```

```python
df4['Year'] = yr
```

```python
df4.head()
```

```python
df4.replace({'Month':{'01':1, '02': 2, '03': 3, '04': 4, '05': 5, '06': 6, '07': 7,
                     '08': 8, '09': 9, '10': 10, '11': 11, '12': 12}}, inplace = True)
```

```python
df4['Month'].value_counts()
```

```python
df4.drop(columns=['Date of Transfer'], inplace=True)
```

```python
df4.head()
```

```python
df4.drop(columns=['Record Status - monthly file only'], inplace=True)
```

```python
missing_value_counts = df4.isnull().sum()
all_data_na = missing_value_counts / len(df4)


#PLOTTING RATIO OF MISSING VALUES FOR EACH COLUMN
plt.subplots(figsize = (19, 4))
sns.barplot(x = all_data_na.index, y = all_data_na)
plt.xlabel('Features')
plt.ylabel('Percentage of missing values')
plt.title('All Columns with missing data for joined dataset')
plt.show()
```
