## Updating Towns and Neighbourhoods
In my earlier data cleaning I took a list of towns and villages in each county and tried to find them within addresses. If they were not found in any way they would be denoted as "X county" and later removed. This was due to the fact these were either incorrectly spelt or just a rural townland with no real predictive power. However, in some of my counties, let's take Cork as an example, I had "Cork" and "Cork City" in my town list. That's because a lot of the addresses would denote the city as simply Cork. However, some of these may contain incorrect spellings of a town or a rural townland containing Cork at the end of the address. Let's say the address was denoted as ***Youghl, Cork***. The way the function was set up was that if it found nothing on the 2nd to last string it would take the last string if it was in our list of towns. As a result some of our values may be faulty with such towns/cities, which typically possess the same name as the county they are from. This would affect the predictive power of our model without doubt. With our column of neighbourhoods now added we can root out these faulty values a bit easier and detect any spelling errors too. This will involve a degree of manual searching online for some townlands/areas but I feel it is a worthwhile experiment to improve our model further.
